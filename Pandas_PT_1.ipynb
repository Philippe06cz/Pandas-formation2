#### concat series

```python
import pandas as pd

python_series_a = pd.Series([1,2,3,4,5], name="Python Series 1", index= [1,2,3,4,5])
python_series_b = pd.Series([6,7,8,9,10], name="Python Series 2", index= [4,5,6,7,8])
```

```python
python_series_a
```

```python
python_series_b
```

```python
new_df_from_two_series = pd.concat([python_series_a, python_series_b], axis=0)
print(new_df_from_two_series)
```

```python
new_df_from_two_series_2 = pd.concat([python_series_a, python_series_b], axis=1)
print(new_df_from_two_series_2)
```

```python
new_df_from_two_series_2.to_excel('wb.xlsx')

excel_df = pd.read_excel('wb.xlsx')
print(excel_df)
```

```python
new_df_from_two_dfs = pd.concat([excel_df, new_df_from_two_series_2], axis=1, keys=["a","b"])
print(new_df_from_two_dfs)
```

# Programmatically Combine Excel Worksheets on Certain Columns

### Programmatically Combine Excel Worksheets on Certain Columns - Five Minute Python Scripts

https://www.youtube.com/watch?v=MqA0ZNP0EDQ

```python
import os
import pandas as pd

data_location = "C:/Users/Philippe/Desktop/ExcelData2/"
desired_headings = ["Année", "Mois", "Valuable Information"]


df_total = pd.DataFrame(columns=desired_headings)


for file in os.listdir(data_location):
    df_file = pd.read_excel(data_location + file)
    selected_columns = df_file.loc[:, desired_headings]

    df_total = pd.concat([selected_columns, df_total], ignore_index=True)


df_total.to_excel("ValuableInformation.xlsx")
```
